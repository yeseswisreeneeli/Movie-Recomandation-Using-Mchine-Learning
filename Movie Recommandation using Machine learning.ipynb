# Movie Recommander Engine


```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_title_from_index(index):
    return dataset[dataset.index == index]["title"].values[0]

def get_index_from_title(title):
    return dataset[dataset.title == title]["index"].values[0]


dataset=pd.read_csv("movie_dataset.csv")
#print(dataset.columns)
features=dataset[["keywords","cast","genres","director"]]
#print(features.head())

for features in features:
    dataset[features]=dataset[features].fillna('')
    
def combined_features(row):
    single_string= row["keywords"]+" "+row["cast"]+" "+row["genres"]+" "+row["director"]
    return single_string

dataset["combined_features"] = dataset.apply(combined_features,axis=1)
#print (dataset["combined_features"].head())


    
##Step 3: Create a column in DF which combines all selected features

##Step 4: Create count matrix from this new combined column

cv = CountVectorizer()

count_matrix=cv.fit_transform(dataset["combined_features"])

##Step 5: Compute the Cosine Similarity based on the count_matrix

similarity_scores = cosine_similarity(count_matrix)
#print(similarity_scores)

movie_user_likes = "Avatar"

## Step 6: Get index of this movie from its title
movie_index= get_index_from_title(movie_user_likes)
#print(movie_index)

## Step 7: Get a list of similar movies in descending order of similarity score
similar_movies=list(enumerate(similarity_scores[movie_index]))

arranged_similar_movies= sorted(similar_movies, key=lambda x:x[1],reverse=True)
                    

## Step 8: Print titles of first 50 movies
i=0
for movie in arranged_similar_movies:
                    print(get_title_from_index(movie[0]))
                    i=i+1
                    if i>50:
                     break
                      
```
